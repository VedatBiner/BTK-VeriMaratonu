### Kütüphanelerin yüklenmesi

```python
from pyspark.sql import SparkSession
```

```python
sparkSession = SparkSession.builder.appName("Veri Maratonu Spark ilk örnek").getOrCreate()
```

```python
sparkSession
```

```python
# veriyi yükleyelim.
sparkSession.read.csv("/FileStore/tables/mock_data_sample-2.csv", header = True).show()
```

```python
# veriyi dataframe'e aktarmak
rawDF = sparkSession.read.csv("/FileStore/tables/mock_data_sample-2.csv", header = True, inferSchema = True)
rawDF.show()
```

```python
# Dosyamızın kolon tipleri
rawDF.printSchema()
```

```python
rawDF.show(truncate = False)
```

```python
rawDF.columns
```

```python
# sadece istediğimiz kolonları görelim
nameCountryOfDF = rawDF.select("first_name", "country")
nameCountryOfDF.show()
```

```python
# kolon bazlı özet
rawDF.describe().show()
```

```python
# yeni kolon ekleme
rawDF.withColumn("new_id", rawDF["id"] + 1000).show()
```

```python
# gruplama
rawDF.groupBy("country").count().show()
```

```python
# Filtreleme
rawDF.filter("country = 'Germany'").show()
```

```python
#  Filtreleme
rawDF.filter("id > 500").show()
```

```python
rawDF.drop("email").show()
```

```python
# sütun adını çıkaralım
rawDF.drop("email").show()
```

```python
# sütun adını değiştirelim.
rawDFwoEmail = rawDF.drop("email")
rawDFwoEmail.withColumnRenamed("first_name", "firstname").show()
```

```python
rawDFwoEmail.withColumnRenamed("last_name", "lastname").show()
```

```python
# Daha güzel bir görünüm için
countByCountryDF = rawDF.select("first_name", "country").groupBy("country").count()
display(countByCountryDF)
```

```python
# Daha güzel bir görünüm için
countByCountryDF = rawDF.select("first_name", "country").groupBy("country").count().orderBy("count")
display(countByCountryDF)
```

```python
# SQL sorgulması için person adında geçici bir view oluşturalım.
rawDF.createOrReplaceTempView("person")
```

```python
%sql
SELECT * FROM person
```

```python
%sql
SELECT first_name, last_name FROM person
```

```python
%sql
SELECT first_name, last_name FROM person GROUP BY country
```

```python
%sql
SELECT first_name, last_name FROM person WHERE country = "Germany"
```
