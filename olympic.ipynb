```python
# kütüphaneler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter

# uyarıları kapatalım.
import warnings
warnings.filterwarnings("ignore")
```

```python
# veriyi okutma
veri = pd.read_csv("olimpiyatlar.csv")
veri.head()
```

```python
# veri bilgileri
veri.info()
```

```python
# sütun isimleri
veri.columns
```

```python
# sütun isimlerini değiştirelim
veri.rename(columns={
    "ID" : "id",
    "Name" : "isim",
    "Sex" : "cinsiyet",
    "Age" : "yas",
    "Height" : "boy",
    "Weight" : "kilo",
    "Team" : "takim",
    "NOC" : "uok",
    "Games" : "oyunlar",
    "Year" : "yil",
    "Season" : "sezon",
    "City" : "sehir",
    "Sport" : "spor",
    "Event" : "etkinlik",
    "Medal" : "madalya"}, inplace =True
)
veri.head(2)
```

```python
# id ve oyunlar sütunlarını çıkaralım
veri = veri.drop(["id", "oyunlar"], axis =1)
veri.head(2)
```

```python
# kayıp verileri dolduralım
# etkinlik 
essiz_etkinlik = pd.unique(veri.etkinlik)
print("Eşsiz Ekinlik sayısı : {}".format(len(essiz_etkinlik)))
essiz_etkinlik[:10]
```

```python
# etkinlikleri dolaşıp, boy ve ağırlık ortalamsını bulalım
# etkinlik özelinde kayıp boy ve kilo değrlerini etkinlik ortalamalarına eşitle
# bu bölüm çalışırken uzun sürüyor.

# gerçek veriyi bozmayalım.
veri_gecici = veri.copy() 

boy_kilo_liste = ["boy", "kilo"]
for e in essiz_etkinlik: # liste içinde dolaş
    # etkinlik filtresi
    etkinlik_filtre = veri_gecici.etkinlik == e
    # etkinliğe göre filtrele
    veri_filtreli = veri_gecici[etkinlik_filtre]

    # boy ve kilo etkinlik özelinde ortalamalar.
    for s in boy_kilo_liste:
        ortalama = np.round(np.mean(veri_filtreli[s]), 2)
        # ~ değil anlamında
        if ~np.isnan(ortalama): # eğer etkinlik özelinde ortalma varsa
            veri_filtreli[s] = veri_filtreli[s].fillna(ortalama)
        else: # eğer etkinlik özelinde ortalma yoksa, ortalamayı hesapla
            tum_veri_ortalamasi = np.round(np.mean(veri[s]), 2)
            veri_filtreli[s] = veri_filtreli[s].fillna(tum_veri_ortalamasi)

    # geçici veriyi gerçek veriye eşitle
    veri_gecici[etkinlik_filtre] = veri_filtreli
    
# kayıp değerleri giderilmiş olan geçici veriyi gerçek veriye eşitle
veri = veri_gecici.copy()
veri.info() # boy ve kilo sütunlarında kayıp veri kaldı mı?

```

```python
# yas değişkeninde tanımlı olmayan değerleri bul
yas_ortalamasi = np.round(np.mean(veri.yas), 2)
print("Yaş Ortalaması : {}".format(yas_ortalamasi))
veri["yas"] = veri["yas"].fillna(yas_ortalamasi)
veri.info()
```

```python
# Madalya alamayanları çıkar
madalya_degiskeni =  veri["madalya"]
pd.isnull(madalya_degiskeni).sum()
```

```python
# madalyası tanımsız olanlar
madalya_degiskeni_filtresi = ~pd.isnull(madalya_degiskeni)
```

```python
# kac kisi kaldı (sadece madalya kazananlar)
veri = veri[madalya_degiskeni_filtresi]
veri.head(5)
```

```python
# kaç kişi kaldı?
veri.info()
```

```python
# sonradan kullanbilmek için veriyi kaydedelim.
veri.to_csv("olimpiyatlar_temizlenmis.csv", index = False)
```

```python
# öncelikli histogramlar çizelim
def plotHistogram(degisken):
    '''
    Girdi : Değişken sütun ismi
    Çıktı : Değişkenin histogram
    '''
    plt.figure()
    plt.hist(veri[degisken], bins = 85, color = "orange")
    plt.xlabel(degisken)
    plt.ylabel("Frekans")
    plt.title("Veri Sıklığı - {}".format(degisken))
    plt.show()
```

```python
# kategorik değişkenler
# çubuk grafiği metodu 
def plotBar(degisken, n = 5):
    '''
    Girdi: Değiken/sütun adı
    n = gösterilecek eşsiz değer sayısı
    Çıktı: Çubuk Grafiği
    '''
    veri_ = veri[degisken]
    veri_sayma = veri_.value_counts()
    veri_sayma = veri_sayma[:n]
    plt.figure()
    plt.bar(veri_sayma.index, veri_sayma, color = "orange")
    plt.xticks(veri_sayma.index, veri_sayma.index.values)
    plt.xticks(rotation = 45)
    plt.ylabel("Frekans")
    plt.title("Veri sıklığı - {}".format(degisken))
    plt.show()
    print("{} : \n {}".format(degisken, veri_sayma))
```

```python
# grafiği çizelim
kategorik_degisken = ["isim", "cinsiyet", "takim", "uok", "sezon", "sehir", "spor", "etkinlik", "madalya"]
for i in kategorik_degisken:
    plotBar(i)
```

```python
# anomali tesbiti ve aykırı değer
def anomaliTespiti(df, ozellik):
    outlier_indices = []

    for c in ozellik:
        # 1. çeyrek
        Q1 = np.percentile(df[c], 25)
        # 3. çeyrek
        Q3 = np.percentile(df[c], 75)
        # IQR hesabı (Inter Quartile Range)
        IQR = Q3 - Q1
        # aykırı değer için ek adım
        outlier_step = 1.5 * IQR
        # aykırı değer ve indeksi
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        # tesbit edilen indeksleri depola
        outlier_indices.extend(outlier_list_col)

    # eşsiz aykırı değerler
    outlier_indices = Counter(outlier_indices)
    # eğer bir örenk v adet sütunda farklı değer ise
    # bunu aykırı kabul edelim.
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 1)

    return multiple_outliers
```

```python
veri_anomali = veri.loc[anomaliTespiti(veri, ["yas", "kilo", "boy"])]
veri_anomali.spor.value_counts()
```
