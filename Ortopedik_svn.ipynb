## SVM (Support Vector Machine) Örneği

```python
# Kütüphaneler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# uyarıları kapat
import warnings
warnings.filterwarnings("ignore")
```

```python
# veriyi oku
data = pd.read_csv("ortopedik_hastalarin_biyomekanik_ozellikleri.csv")
data.head()
```

```python
# class verilerini sayısala çevirelim 
data["class"] = [1 if each == "Abnormal" else 0 for each in data["class"]]
data.head()
```

```python
# class sütununu çıkaralım. Sınıfları y içine koyalım.
# y sadece etiketlerden oluşsun
y = data["class"].values
x_data = data.drop(["class"], axis = 1) 
```

```python
# normalizasyon (min-max)
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
```

```python
# Eğitim test bölünmesi kütüphanesi
from sklearn.model_selection import train_test_split
```

```python
# eğitim test ayrılması
# %15 test, %85 Eğitim
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 1)
```

```python
# SVM Kütüphanesi
from sklearn.svm import SVC
```

```python
# SVM eğitimi
svm = SVC(random_state = 1)
svm.fit(x_train, y_train)

# testi
print("Destek Vektör Makinesi Modeli Test Doğruluk : {} ".format(svm.score(x_test, y_test)))
```
