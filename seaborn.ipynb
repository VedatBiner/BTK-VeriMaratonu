## Seaborn Örnekleri

### Kütüphanelerin yüklenmesi

```python
# kütüphaneler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# uyarıları kapatalım.
import warnings
warnings.filterwarnings("ignore")
```

### Verilerin Okunması

```python
# veriyi oku
veri = pd.read_csv("/DATA/datasets/olimpiyatlar_temizlenmis.csv")
veri.head()
```

### Saçılım Grafiği

```python
# saçılım grafiği
def sacilimGrafikCiz():
    sns.scatterplot(x = "boy", y = "kilo", data = veri)
    plt.title("Boy ve Ağırlık İlişkisi - Beyaz Izgara Tema")
    plt.show()

sacilimGrafikCiz()
```

```python
# koyu tema
sns.set_style("darkgrid")
sacilimGrafikCiz()
```

```python
sns.set_style("whitegrid")
sacilimGrafikCiz()
```

### Saçılım Grafiği - Madalyaya Göre

```python
# Saçılım Grafiği - Kategorik - Madalyaya göre
sns.scatterplot(x = "boy", y = "kilo", hue = "madalya", data = veri)
plt.title("Madalyaya Göre Boy ve Ağırlık Dağılımı")
plt.show()
```

### Saçılım Grafiği Cinsiyete Göre

```python
# Cinsiyete göre
sns.scatterplot(x = "boy", y = "kilo", hue = "cinsiyet", data = veri)
plt.xlabel("Sporcuların boyu")
plt.ylabel("sporcuların ağırlığı")
plt.title("Cinsiyete Göre Boy ve Ağırlık Dağılımı")
plt.show()
```

```python
# Cinsiyete göre
sns.set_style("whitegrid")
sns.scatterplot(x = "boy", y = "kilo", hue = "cinsiyet", style = "madalya", data = veri)
plt.xlabel("Sporcuların boyu")
plt.ylabel("sporcuların ağırlığı")
plt.title("Madalyalı Sporcuların Ağırlık-Boy Değişimleri")
plt.show()
```

### Yaşa Göre İmleç Belirleyelim

```python
# İmleçleri yaşa göre belirtelim.
sns.set_style("whitegrid")
sns.scatterplot(x = "boy", y = "kilo", hue = "cinsiyet", style = "madalya", 
    size = "yas",  sizes = (15, 200), data = veri)
plt.xlabel("Sporcuların boyu")
plt.ylabel("sporcuların ağırlığı")
plt.title("Madalyalı Sporcuların Ağırlık-Boy Değişimleri")
plt.show()
```

### Saçılım Gragfiği - Doğrusal Regresyon

```python
# saçılım grafiği - doğrusal regresyon
sns.regplot(x = "boy", y = "kilo", data = veri, marker = "+", scatter_kws={"alpha" : 0.2})
plt.title("Boy ve ağırlık Dağılımı")
plt.show()
```

### Renk Paleti Uygulayalım

```python
# renk paleti
sns.scatterplot(x = "boy", y = "kilo", hue = "madalya", data = veri, palette = ("Set1"))
plt.title("Madalyaya Göre Boy ve ağırlık Dağılımı")
plt.show()
```

```python
# renk paleti - 2
sns.scatterplot(x = "boy", y = "kilo", hue = "madalya", data = veri, palette = ("rocket"))
plt.title("Madalyaya Göre Boy ve ağırlık Dağılımı")
plt.show()
```

```python
# çizgi grafiği
sns.lineplot(x = "boy", y = "kilo", data = veri)
plt.title("Boy ve Kilo")
plt.show()
```

### Histogramlar

```python
# histogram
sns.displot(veri, x = "kilo")
plt.ylabel("Frekans")
plt.title("Ağırlık Histogramı")
plt.show()
```

```python
# cinsiyete göre histogram
sns.displot(veri, x = "kilo", hue = "cinsiyet")
plt.ylabel("Frekans")
plt.title("Cinsiyete göre ayrıştırılmış histogram")
plt.show()
```

```python
# cinsiyete göre histogram
sns.displot(veri, x = "kilo", hue = "cinsiyet")
plt.ylabel("Frekans")
plt.title("Cinsiyete göre ayrıştırılmış histogram")
plt.show()
```

```python
# Histogram - Cinsiyet
sns.displot(veri, x = "kilo", col = "cinsiyet", multiple = "dodge")
plt.show()
```

```python
# İki boyutlu Histogram - ok.
sns.displot(veri, x = "kilo", y = "boy", kind = "kde")
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("Ağırlık-Boy Histogramı")
plt.show()
```

### İki Boyutlu Histogram

```python
# İki boyutlu Histogram
sns.displot(veri, x = "kilo", y = "boy", hue = "cinsiyet", kind = "kde")
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("cinsiyete Göre Ayrıştırılmış Ağırlık-Boy Histogramı")
plt.show()
```

```python
# İki boyutlu Histogram 
sns.kdeplot(data = veri, x = "kilo", y = "boy", hue = "cinsiyet", fill = True)
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("cinsiyete Göre Ayrıştırılmış Ağırlık-Boy Histogramı")
plt.show()
```

### Çubuk Grafikler

```python
# Çubuk grafiği
sns.barplot(x = "madalya", y = "boy", data = veri)
plt.title("Madalyalara Göre Boy Grafikleri")
plt.show()
```

```python
# Çubuk grafiği
sns.barplot(x = "madalya", y = "boy", data = veri, hue = "cinsiyet")
plt.title("Madalyalara Göre Boy Grafikleri")
plt.show()
```

```python
# Kategorize ederek çizelim
sns.catplot(x = "madalya", y = "yas", hue = "cinsiyet", col = "sezon", data = veri, 
    kind = "bar", height = 4, aspect = 0.7)
plt.show()
```

### Box Grafik

```python
# box grafik
sns.boxplot(x = "sezon", y = "boy", data = veri)
plt.show()
```

```python
# box grafik
sns.boxplot(x = "sezon", y = "boy", hue = "cinsiyet", data = veri, palette = "Set1")
plt.show()
```

```python
# bir örnek daha
veri_gecici = veri.loc[:,["yas", "kilo", "boy"]]
sns.boxplot(data = veri_gecici, orient = "h", palette = "Set2")
plt.show()
```

### Catplot

```python
# catplot örneği
sns.catplot(x = "sezon", y = "boy", hue = "cinsiyet", col = "madalya", 
    data = veri, kind = "box", height = 6, aspect = 0.7)
plt.show()
```

### Ortaklık Grafiği

```python
# ortaklık grafiği
sns.jointplot(x = "kilo", y = "boy", hue = "sezon", kind = "kde", data = veri)
plt.show()
```

### Ortaklık, Box, Histogram Bir arada

```python
# Ortaklık, histogram ve box grafiği bir arada
g = sns.JointGrid(data = veri, x = "kilo", y = "boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

### Pair Plot

```python
# pair plot (eş grafiği)
sns.pairplot(veri)
plt.show()
```

```python
# pair graph - 2
# burasını çalışması yaklaşık 6 dak. sürebiliyor.
g = sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill = True)
g.map_diag(sns.histplot, kde  =True)
plt.show()
```

### Sayma Grafiği

```python
# sayma grafiği
sns.countplot(x = "sehir", data = veri)
plt.xticks(rotation = 90)
plt.show()
```
